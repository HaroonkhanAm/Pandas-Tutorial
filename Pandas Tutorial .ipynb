Series is a one-dimensional array capable of holding data of all type (integer, string, float, python objects, etc.)
A pandas Series has no column labels, as it is just a single column of a DataFrame . A Series does have row labels.

### Access elements using index labels

```python
my_list = ["Yes", "No", 1, 10]
indices = ['Orange', 'Peach', 'Mango', 'Grapes']

Fruits = pd.Series(data = my_list, index = indices)
Fruits
```

```python
print(Fruits)
print()

# Access elements in Fruits using index labels:

# Use of a single index label
print("Fruits['Peach']:\n", Fruits['Peach'])
print()

# Access multiple index labels
print("Fruits[['Orange', 'Peach']]:\n", Fruits[['Orange', 'Peach']])
print()

# Use loc to access multiple index labels
print("Fruits.loc['Grapes', 'Mango']:\n", Fruits.loc[['Grapes', 'Mango']])
print()

# Access elements in Fruits using numerical indices:

# Use multiple numerical indices
print("Fruits.[[0, 1]]:\n", Fruits[[0, 1]])
print()

# Use a negative numerical index
print("Fruits[[-1]]:\n", Fruits[[-1]])
print()

# Use a single numerical index
print("Fruits[0]:\n", Fruits[0])
print()

# Use iloc to access multiple numerical indices
print("Fruits.iloc[[2, 3]]:\n", Fruits.iloc[[2, 3]])
```

### Deleting element of  a Series

```python
print("Before deleting:\n")
print(Fruits)
print()

Fruits.drop('Peach', inplace = True)

print("After deleting:\n")
print(Fruits)
```

### Binary Operations on Pandas Series

```python
Veg = pd.Series(data = [6, 18, 33], index = ["Tomatoes", "Onion", "Potatoes"])
print(Veg)
print()

print("Veg + 2:\n", Veg + 2)
print()

print("Veg - 2:\n", Veg - 2)
print()

print("Veg * 2:\n", Veg * 2)
print()

print("Veg / 2:\n", Veg / 2)
```

```python
df1 = pd.DataFrame([[1, 2], [3, 4]], columns=list('AB'), index=['x', 'y'])   #Create a dataframe 1 for append method.
df2 = pd.DataFrame([[5, 6], [7, 8]], columns=list('AB'), index=['x', 'y'])   #Create a dataframe 2 for append method.
df1.append(df2, ignore_index=True)                  #  Combine 2 dataframe with help of append() method.
```

```python
dataf1 = pd.DataFrame({'key1': ['A', 'B', 'C', 'D'],
                     'value': [1, 2, 3, 5]})
dataf2 = pd.DataFrame({'key2': ['A', 'B', 'C', 'D'],        #Merging DataFrames object with eachother
                    'value': [5, 6, 7, 8]})
dataf1.merge(dataf2, left_on='key1', right_on='key2')
```

```python
Series1 = pd.Series(['a', 'b'])
Series2 = pd.Series(['c', 'd'])       # Using the cancat() method to combine series
pd.concat([Series1, Series2], ignore_index=True)
```

```python
Color1= pd.DataFrame([['Blue',1], ['Red, 2'],['Green',3]])
Color2= pd.DataFrame([['Orange',4], ['Purple, 5'],['Yellow',6]]) # Using the cancat() method to combine DataFrames
pd.concat([Color1,Color2])
```
